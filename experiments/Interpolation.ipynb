```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

```python
x = [0, 50, 100, 300, 500, 600, 700, 800, 900, 1000]
y = [0.1, 0.1, 0.1, 0.1, 5, 10, 18, 35, 42, 60]

x = [0, 250, 500, 600, 700, 800, 900, 1000]
y = [0.1, 0.5, 1, 7, 15, 26, 42, 60]

plt.plot(x, y)
```

```python
# cf. https://pythonnumericalmethods.berkeley.edu/notebooks/chapter17.03-Cubic-Spline-Interpolation.html
```

```python
# use bc_type = 'natural' adds the constraints as we described above
cubic_spline = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 1000, 100)
y_new = cubic_spline(x_new)
```

```python
plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
cubic_spline(950)
```
